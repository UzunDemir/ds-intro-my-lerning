```python
# Импортируйте необходимые библиотеки
import requests
import pandas as pd
```

**Задача 1**

```python
# 1.
url = 'https://itunes.apple.com/lookup'
params = {'id': '1425235648'}   
response = requests.get(url, params=params)
response_json = response.json()
response_json
```

```python
# 2. 
response_json['resultCount']
```

```python
# 3.
response_json['results'][0]['primaryGenreName']
```

**Задача 2**

```python
# 1.
url = 'https://itunes.apple.com/search'
params = {'term': 'Einaudi', 'limit': 150, 'media': 'music'}   
response = requests.post(url, data=params)
response_json = response.json()
response_json
```

```python
# 2.
df = pd.DataFrame(response_json['results'])
df
```

```python
df.shape()
```

```python
# 3.

df.to_excel('data/Ludovico Einaudi.xlsx', sheet_name='track', index_label='id')
```

```python
# 4.

df.to_csv('data/Ludovico Einaudi.csv', sep='|')
```
